### **Image to Pencil Sketch**

```python
# Importing necessary libraries for our task.

import io 
import cv2 # OpenCV
from PIL import Image
from IPython.display import display
import matplotlib.pyplot as plt # For interpreting data in graphs
%matplotlib inline
```

### Uploading an Image

```python
img = cv2.imread('wonder-day-among-us-png-31.png')
```

### Resizing the Image based on our Requirements.

```python
img = cv2.resize(img, (600, 500))
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

### Converting the image to Gray Scale

```python
new_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(new_image))
```

### Inverting the image

```python
invert = cv2.bitwise_not(new_image)
display(Image.fromarray(invert))
```

### Removing Noise in the image

```python
blur_image = cv2.GaussianBlur(invert, (5, 5), 0, 0)
display(Image.fromarray(blur_image))
```

### Now, we have our pencil sketch

```python
invert_blur_image = 255 - blur_image
pencil_sketch = cv2.divide(new_image, invert_blur_image, scale=250.0)
display(Image.fromarray(pencil_sketch))
```
